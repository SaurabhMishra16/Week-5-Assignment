Q3. What errors are defined in theArithmeticError class? Explain any two with an example.

There are 3 types of errors are defined in the ArithmeticError class
1- ZeroDivisionError
2- OverFlowError
3- FloatingPointError

```python
# ZeroDivisionError
import logging
logging.basicConfig(filename = 'record.txt', level = logging.DEBUG)
# try block generate error
try:
    a = 12
    b = 0
    a/b
# except catch the error
except ZeroDivisionError as e:
    print('First ArithmeticError is ',e)
    logging.info(f'First ArithmeticError is {e}')
```

```python
#FloatingPointError
try:
    result = 1.0 / 0.0  # Division by zero
except FloatingPointError as e :
    print("Floating point error occurred." , e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is built-in Exception class in python that serves as the base class for exception related to lookup operation

Lookup operations involve searching for or accessing elements in data structures like lists, dictionaries, or sets. The LookupError class provides a common base for exceptions that can occur during these lookup operations.

```python
# Generating KeyError
import logging
list= {'name':'saurabh',
      'age':23,
      'course':'Data Science Master'}

logging.basicConfig(filename = 'record.txt', level = logging.NOTSET)

try:
    logging.info(list['address'])
except KeyError as ke:
    logging.info('This is generating KeyError')
```

```python
# Generating IndexError
import logging
list= [1,2,3,4,5]

logging.basicConfig(filename = 'record.txt', level = logging.NOTSET)

try:
    logging.info(list[9])
except IndexError as ie:
    logging.info('This is generating IndexError')
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError:- A ImportError class is built-in Exception class in python that is raised when an import statement fails to find and load module or when there is an error in the process of importing a module

ModuleNotFoundError:- A ModuleNotFoundError exception class was introduced as in subclass of 'ImportError' class. it specifically handle cases where a module cannot be found during the import process.
